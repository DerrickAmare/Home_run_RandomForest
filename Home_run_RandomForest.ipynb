```python
import pandas as pd
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
df = pd.read_csv(list(uploaded.keys())[0], encoding = 'latin1')
```

```python
df = df.drop(columns=['game_date','player_name','des','batter','pitcher','home_team','away_team'])
```

```python
df = df.dropna()
display(df.head())
```

```python
df = pd.get_dummies(df, columns=['pitch_name'], drop_first=True)
display(df)
```

```python
# Drop 'events' and other non-numerical columns from the features (X)
X = df.drop(columns=['events', 'description', 'Batter_position', 'Pitcher_position', 'bb_type'])

# The target variable (y) remains the 'events' column
y = df.loc[:,'events']
```

```python
from  sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 11)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
prediction = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
features = rf.feature_importances_
```

```python
features = pd.DataFrame(rf.feature_importances_, index =X.columns, columns=['Importance'])
```

```python
display(features.sort_values(by=['Importance'], ascending=False))
```
